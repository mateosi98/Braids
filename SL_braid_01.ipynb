```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from ast import literal_eval 
```

```python
dir_name = "/Users/mateosallesize/Google Drive/SRO/Braids/Supervised/Data/"
strands = 4
length = 6
depth = 5
file_name = "braid_tree_l"+str(length)+"_s"+str(strands)+"_d"+str(depth)+".csv"
```

```python
(strands*2+4)*length
```

```python
data = pd.read_csv(dir_name+file_name, converters={'braid':literal_eval,'labels':pd.eval}).drop("Unnamed: 0", axis=1)
msk = np.random.rand(len(data)) < 0.8
training_data = data[msk]
training_braids = training_data['braid'].values.tolist()
training_labels = training_data['labels'].values.tolist()
labels = []
for label in training_labels:
    labels.append(label[0])
training_labels = labels
test_data = data[~msk]
test_braids = test_data['braid'].values.tolist()
test_labels = test_data['labels'].values.tolist()
labels = []
for label in test_labels:
    labels.append(label[0])
test_labels = labels
```

```python
print('The unique labels that we have in the training labels',set(training_labels))
```

```python
training_braids
```

```python
tf.keras.backend.clear_session()
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense(64, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense(32, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense((strands*2+4)*length, activation=tf.nn.softmax)])
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(training_braids, training_labels, epochs=100)
```

```python
model.evaluate(test_braids, test_labels)
```
